```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Porsche Hackathon Data - Sheet1.csv")
```

```python
df.head()
```

```python
df['MSRP'] = [x.replace('$','') for x in df['MSRP']] 
df['MSRP'] = [x.replace(',','') for x in df['MSRP']]
df['MSRP']=pd.to_numeric(df['MSRP'],errors='coerce')
df['36 months'] = [x.replace('$','') for x in df['36 months']]
df['36 months'] = [x.replace(',','') for x in df['36 months']]
df['36 months']=pd.to_numeric(df['36 months'],errors='coerce')
df['48 months'] = [x.replace('$','') for x in df['48 months']]
df['48 months'] = [x.replace(',','') for x in df['48 months']]
df['48 months']=pd.to_numeric(df['48 months'],errors='coerce')
df['60 months'] = [x.replace('$','') for x in df['60 months']]
df['60 months'] = [x.replace(',','') for x in df['60 months']]
df['60 months']=pd.to_numeric(df['60 months'],errors='coerce')
df['Price/L'] = [x.replace('$','') for x in df['Price/L']]
df['Price/L'] = [x.replace(',','') for x in df['Price/L']]
df['Price/L']=pd.to_numeric(df['Price/L'],errors='coerce')
```
